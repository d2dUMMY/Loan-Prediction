```python
train.drop('Loan_ID', axis=1, inplace=True)
id = test.Loan_ID
test.drop('Loan_ID', axis=1, inplace=True)
```

```python
test.head()
```

```python
train.head()
```

```python
from sklearn.preprocessing import LabelEncoder
columns = ['Gender','Married','Education','Self_Employed','Property_Area','Dependents']

for col in columns:
    le = LabelEncoder()
    train[col] = le.fit_transform(train[col])
    test[col] = le.transform(test[col])
```

```python
train.head()
```

```python
test.head()
```

```python
le = LabelEncoder()
train.Loan_Status = le.fit_transform(train.Loan_Status)
train.head()
```

```python
from sklearn.preprocessing import StandardScaler

columns = ['ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term']

for col in columns:
    scale = StandardScaler()
    train[col] = pd.Series(scale.fit_transform(train[col].values.reshape(-1,1)).reshape(-1,))
    test[col] = pd.Series(scale.transform(test[col].values.reshape(-1,1)).reshape(-1,))
```

```python
#fitting RandomForest classifier to the dataset
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(train.drop('Loan_Status', axis=1), train.Loan_Status)
```

```python
ans= rf.predict(test)
```

```python
ans = le.inverse_transform(ans)
```

```python
pd.DataFrame(data = {"Loan_ID":id, "Loan_Status":ans})
result = pd.DataFrame(data = {"Loan_ID":id, "Loan_Status":ans})
print("Writing output file to rf.csv")
result.to_csv("./dataset/rf.csv",index=False)
```

```python
from sklearn.svm import SVC

svc = SVC()
svc.fit(train.drop('Loan_Status', axis=1), train.Loan_Status)
```

```python
ans= svc.predict(test)
```

```python
ans=le.inverse_transform(ans)
```
